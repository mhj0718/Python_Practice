```python
import pandas as pd
df = pd.read_csv('c:/py_data/파이썬지식인 크롤링.csv', encoding='cp949')
text_all=''
for i in df.index:
    text_all = text_all + " " +df.iloc[i][2]    
text_all
```

```python
Twitter.add_dictionary('판다스','Noun')
```

```python
Tpos = Twitter.pos(text_all)
Tpos
```

```python
# 형용사만 분류
adj_lst = []
for tag in Tpos:
    if tag[1] in ['Adjective']:
        adj_lst.append(tag[0])

adj_lst
```

```python
# 형용사, 명사 분류
adj_noun_lst = []
for tag in Tpos:
    if tag[1] in ['Adjective', 'Noun']:
        adj_noun_lst.append(tag[0])

adj_noun_lst
```
