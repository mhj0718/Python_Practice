```python
url1 = 'http://api.visitkorea.or.kr/openapi/service/rest/KorService/searchStay?serviceKey='
key = 'c1N7z7KEDvTRSI1eJtuHGiVHaOOlF7yuSjsMNvcd5CsskIzYTNbPuki3ImWbHToiVxBnGacopQcmmZrOjpohPg%3D%3D'
url2 = '&numOfRows=43&pageSize=10&pageNo=1&MobileOS=ETC&MobileApp=AppTest&arrange=P&listYN=Y&benikia=1'
url = url1 + key + url2
print(url)
```

```python
xml = requests.get(url)
soup = BeautifulSoup(xml.text, 'html.parser')
```

```python
data_lst = soup.find_all('item')
```

```python
data_lst
```

```python
len(data_lst)
```

```python
test_lst = []
for i in data_lst:
    ADD_1 = i.find('addr1').text
    TITLE = i.find('title').text
    READ_CNT = i.find('readcount').text
    
    test_lst.append({'베네키아숙소이름':TITLE, '주소':ADD_1,
                     '조회수':READ_CNT})
    
test_lst 
    
```

```python
data_lst = pd.DataFrame(test_lst)
data_lst
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm    # 파일 경로 지시
fl = fm.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
plt.rc('font',family=fl)

data_lst.set_index('베네키아숙소이름', inplace=True)
data_lst
```

```python
data_lst['조회수'] = data_lst['조회수'].astype(int)
data_lst['조회수'].plot(kind='barh', title = '숙소별 조회수',
                        figsize=(10,10))
```
