```python
title_lst = soup.find_all('div', class_ = 'tit3')
title_lst
```

```python
name_lst = []
for i in title_lst:
    temp = i.find('a').text
    rank_lst.append(temp)
    
name_lst
```

```python
name_lst = []
for i in range(len(title_lst)):
    temp = title_lst[i].find('a').text
    name_lst.append(temp)
name_lst
```

```python
rank_lst = []
name_lst = []
rank = 0
for i in range(len(title_lst)):
    temp = title_lst[i].find('a').text
    rank += 1
    name_lst.append(temp)
    rank_lst.append(rank)
```

```python
movie_df = pd.DataFrame({'순위':rank_lst,
                        '영화제목':name_lst})
movie_df.set_index('순위', inplace=True)
movie_df
```
