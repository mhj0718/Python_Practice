```python
from bs4 import BeautifulSoup
from urllib.parse import quote
from urllib.request import urlopen
import pandas as pd
import re
import time

keyword = quote(input('검색어를 입력하세요: '))
start_page = int(input('시작 페이지 번호: '))
end_page = int(input('마지막 페이지 번호: '))+1
main_url = 'https://top.cafe.daum.net/_c21_/search?search_opt=board&SearchType=tab&sort_type=accuracy&q='
num_url = '&p='
cafe_url = 'http://cafe.daum.net'
tit_lst=[]
doc_lst=[]
comit_lst=[]
for page_num in range(start_page, end_page):
    url = main_url + keyword + num_url + str(page_num)
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')
    
    link_url_class = soup.find_all('a', class_='link_url')
    suburl_lst = []
    for i in link_url_class:
        suburl = i['href']
        suburl_lst.append(suburl)
        
    for nowurl in suburl_lst:
        now_html = urlopen(nowurl)
        now_soup = BeautifulSoup(now_html, 'html.parser')
        
        src = now_soup.find('iframe', id= 'down')['src']
        
        iframe_html = urlopen(cafe_url+src)
        iframe_soup = BeautifulSoup(iframe_html,'html.parser')
        try:
            title = iframe_soup.find('strong', class_='tit_info').text
            content = iframe_soup.find('xmp', id='template_xmp').text
        except AttributeError:
            continue
            
        comment_lst = iframe_soup.find_all('span', class_='original_comment')
        comment = ''
        for i in comment_lst:
            comment = comment + i.text
       
        tit_lst.append(title)
        doc_lst.append(content)
        comit_lst.append(comment)
        
daum_iframe_df = pd.DataFrame({'제목':tit_lst, '본문':doc_lst, '댓글':comit_lst})
daum_iframe_df.to_csv('c:/py_data/다음카페+'+keyword+'iframe 크롤링 저장.csv')
        
```
