```python
# 함수로 만들기
SentiWord = pd.read_json('c:/py_data/SentiWord_info.json', encoding='utf-8')
def neg_pos(word):
    temp = pd.DataFrame()
    global SentiWord
    temp = SentiWord[(SentiWord['word'] == word)
          | (SentiWord['word_root'] == word)]
    try:
        word_tul = (word,temp['polarity'][temp.index[0]])
    except IndexError:
        word_tul =(word,0)
    return word_tul

```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
##################################
#                                #
#          형태소 분석           #
#                                #
##################################

# KoNLPy는 다음과 같은 다양한 형태소 분석, 
# 태깅 라이브러리를 파이썬에서 쉽게 사용할 수 
# 있도록 모아놓았다.

# Hannanum: 한나눔. KAIST Semantic Web Research Center 개발.
# http://semanticweb.kaist.ac.kr/hannanum/

# Kkma: 꼬꼬마. 서울대학교 IDS(Intelligent Data Systems) 연구실 개발.
# http://kkma.snu.ac.kr/

# Komoran: 코모란. Shineware에서 개발.
# https://github.com/shin285/KOMORAN

# Mecab: 메카브. 일본어용 형태소 분석기를 한국어를 사용할 수 있도록 수정.
# https://bitbucket.org/eunjeon/mecab-ko

# Open Korean Text: 오픈 소스 한국어 분석기. (구.트위터)
# 과거 트위터 형태소 분석기.


# nouns : 명사 추출
# morphs : 형태소 추출
# pos : 품사 부착
```

```python
m = okt.morphs(text_all, stem=True)   # 형태소 추출
```

```python
m
```

```python
from tqdm import tqdm
neg_lst = []
pos_lst = []
unkwon_lst = []
for i in tqdm(m, '긍/부정어를 나누고 있습니다.'):
    result = neg_pos(i)
    if result[1] > 0:
        pos_lst.append(result[0])
    elif result[1] < 0:
        neg_lst.append(result[0])
    elif result[1] == 0:
        unkwon_lst.append(result[0])
```

```python
print('긍정키워드 수:', len(pos_lst),'개')
print('부정키워드 수:', len(neg_lst),'개')
print('중립 or Unknown 키워드 수:', len(unkwon_lst),'개')
```

```python
unkwon_lst
```

```python
from collections import Counter     # jdk에서 제공하는 라이브러리
unkwon_count = dict(Counter(unkwon_lst).most_common())   # 많이 나온순서대로 내림차순 정렬
unkwon_count
```

```python
from collections import Counter             
pos_count = dict(Counter(pos_lst).most_common())
pos_count
```

```python
neg_count = dict(Counter(neg_lst).most_common())
neg_count
```

```python
stopword = ['파이썬','수','좀','제','요','것','뭐','줄',
            '생','를','판','때','하나요','로','곳','비',
            '이','분','문','게','중','저','이제','쪽',
            '거','개','아시','생각','가요','이야기','성기',
            '한게임','해메','판다','스랑','그','더','안','후','뭘','번','걸','위','해']
temp_stopword = ['자','살','상']
for i in temp_stopword:
    stopword.append(i)# 딕셔너리에 key값에 stop_word의 i값을 배분하여 해당하는 value를 가진 item을 지운다
```

```python
temp_dic = {}
for key, value in pos_count.items():
    if key not in stopword:
        temp_dic[key] = value
pos_count = temp_dic
```

```python
temp_dic={}
for key, value in neg_count.items():
    if key not in stopword:
        temp_dic[key] = value
neg_count = temp_dic        
```
