```python
start = int(input('오늘은 어디부터?'))
end = int(input('어디까지?'))

url_df = pd.read_csv('c:/py_data/20210603인스타그램+등산서브url.csv')
sub_url_lst = list(url_df['url'])
sub_url_lst

id_ = '01026778751'
pw= 'agw2539nhh!'

driver = webdriver.Chrome('c:/py_data/chromedriver.exe')
driver.get('https://www.instagram.com/')
time.sleep(2)

# 로그인
id_input = driver.find_element_by_css_selector('#loginForm > div > div:nth-child(1) > div > label > input')
id_input.send_keys(id_)
pw_input = driver.find_element_by_css_selector('#loginForm > div > div:nth-child(2) > div > label > input')
pw_input.send_keys(pw)
driver.find_element_by_css_selector('#loginForm > div > div:nth-child(3) > button > div').click()
time.sleep(5)
```

```python
hash_tag_lst=[]
cont_lst=[]
for url in tqdm(sub_url_lst[start:end], '게시글 크롤링'):
    driver.get(url)
    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    # 해시태그
    class_EtaWk = soup.find('div',class_='EtaWk')
    one_hash_tag=[]
    tags_lst = class_EtaWk.find_all('a',class_=' xil3i')
    for i in tags_lst:
        one_hash_tag.append(i.text)
    
    # 본문
    cont = soup.find('div', class_='C4VMK').text
    cont = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣!?#]',"",cont)
    
    hash_tag_lst.append(one_hash_tag)
    cont_lst.append(cont)
df = pd.DataFrame({'본문':cont_lst,'해시태그':hash_tag_lst})
driver.close()
df.to_csv('c:/py_data/인스타게시글'+str(start)+'~'+str(end)+'까지.csv')
```
