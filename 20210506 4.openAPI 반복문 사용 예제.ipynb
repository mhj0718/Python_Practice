```python
client_id = 'UWHF47Ojv3VPR1mQx_xa'
client_secret = 'kvL2c2dFMA'

keyword = input('네이버 뉴스 검색어 입력: ')
headers = {'X-Naver-Client-Id': client_id,
           'X-Naver-Client-Secret': client_secret}

base_url ='https://openapi.naver.com/v1/search/news?query='
start_url = '&start='
dis_url = '&display='

RANGE = range(1,51)
test_lst = []
for i in tqdm(RANGE):
    url = base_url + keyword + start_url + str(i) + dis_url + str(100)
    response_data = requests.get(url, headers=headers)
    if response_data.status_code == 200:
        temp_data = response_data.json()
        test_lst.extend(temp_data['items'])
        time.sleep(0.1)
    else:
        print(url)
        print(response_data, '응답 받지 못함')


df = pd.DataFrame(test_lst)
df.to_csv('c:/py_data/네이버openAPI뉴스'+keyword+'저장.csv')

```
