```python
tit_lst =[]
cont_lst = []
comm_lst = []
err_lst = []

driver = webdriver.Chrome('c:/py_data/chromedriver.exe')
cafe_main_url = 'https://m.cafe.naver.com/ArticleSearchList.nhn?search.query='
keyword = '분할'
cafe_end_url = '&search.menuid=0&search.searchBy=0&search.sortBy=date&search.clubid=11100373&search.option=0&search.defaultValue=1'
url = cafe_main_url + keyword + cafe_end_url
driver.get(url)

while True:
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    now_num = int(soup.find('span', class_='cnt').text)
    end_num = int(soup.find('span', class_='total').text)
    if now_num >= end_num:
        break
    else:
        driver.find_element_by_css_selector('#moreButton').click()
        time.sleep(2)
        
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
section_class = soup.find('div', class_='search_list')
li_tag_lst = section_class.find_all('li')
 
for i in range(0,20):
    url = 'https://m.cafe.naver.com' + li_tag_lst[i].find('a')['href']
    driver.get(url)
    time.sleep(1)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    try:
        title = soup.find('h2', class_ = 'tit').text
    except:
        continue
    
    try:
        cont = soup.find('div', class_ = 'content').text
    except:
        cont = ''
    
    try:
        comm = soup.find('ul', class_ = 'CommentList').text
    except:
        comm = ''
    
    title = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣]',' ',title)
    tit_lst.append(title)
    cont = re.sub('[^ㄱ-ㅣ가-힣]',' ',cont)
    cont_lst.append(cont)
    comm = re.sub('[^ㄱ-ㅣ가-힣]',' ',comm)
    comm_lst.append(comm)
    
df = pd.DataFrame({'제목':tit_lst,'내용':cont_lst, '댓글':comm_lst})
df.to_csv('c:/py_data/잠백이 헬스카페 크롤링.csv', encoding='cp949')
df


# 모범답안
# tit_lst = []
# doc_lst = []
# cmt_lst = []

# main_url = 'https://m.cafe.naver.com'
# for urlnum in li_tag_lst:
#     now_url = main_url + urlnum.find('a')['href']
#     driver.get(now_url)
#     time.sleep(2)
#     sub_html = driver.page_source
#     sub_soup = BeautifulSoup(sub_html, "html.parser")
    
#     #타이틀
#     try:
#         title = sub_soup.find('h2', class_='tit').text
#     except:
#         continue
#     title = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣]', "", title)
#     tit_lst.append(title)
    
#     #본문
#     try:
#         doc = sub_soup.find('div', class_ = 'se-component-content').text
#     except:
#         doc = ''
#     doc = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣]', "", doc)
#     doc_lst.append(doc)
    
#     #댓글
#     try:
#         CommentList_class = sub_soup.find('ul', class_ = 'CommentList')
#         CommentList = CommentList_class.find_all('p')
#         Comment = ""
#         for Com in CommentList:
#             Comment = Comment + Com.text
#     except:
#         Comment = ""
#     Comment = re.sub('[^0-9a-zA-Zㄱ-ㅣ가-힣]', "", Comment)
#     cmt_lst.append(Comment)
    
# driver.close()
# df = pd.DataFrame({"제목" : tit_lst, "본문":doc_lst, "댓글":cmt_lst})
# df.to_csv("c:/py_data/"+keyword+"카페 로그인 크롤링 저장.csv", encoding="")
```
