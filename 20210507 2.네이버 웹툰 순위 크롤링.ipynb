```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://comic.naver.com/index.nhn'
```

```python
html_url = requests.get(url).text
html_url
```

```python
soup = BeautifulSoup(html_url, 'html.parser')
soup
```

```python
soup.find_all('li', class_ = 'rank01')[0].find('a').text
```

```python
name_lst = []
rank_lst = []
rank = 0
for i in range(1,11):
    if i < 10:
        temp = soup.find_all('li', class_ = 'rank0'+str(i))[0].find('a').text
        name_lst.append(temp)
        rank += 1
        rank_lst.append(rank)
    elif i == 10:
        temp = soup.find_all('li', class_ = 'rank'+str(i))[0].find('a').text
        name_lst.append(temp)
        rank += 1
        rank_lst.append(rank)
name_lst
```

```python
rank_lst
```

```python
weptoon_df = pd.DataFrame({'순위':rank_lst,
                           '웹툰제목':name_lst})
weptoon_df.set_index('순위', inplace=True)
weptoon_df
```

```python
weptoon_df.to_csv('c:/py_data/20210507 네이버웹툰 순위.csv', encoding='cp949')
```
