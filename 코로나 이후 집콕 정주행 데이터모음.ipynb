```python
import pandas as pd
df = pd.read_csv('c:/py_data/py_project/코로나 이후 집콕.csv', encoding='cp949')
df.drop(df.index[2], inplace=True)
del df['Unnamed: 0']
df
```

```python
df_1 = pd.read_csv('c:/py_data/py_project/구글 뉴스-코로나 이후 집콕.csv', encoding='cp949')
del df_1['Unnamed: 0']
df_1.drop(df_1.index[1], inplace=True)
df_1
```

```python
df_2 = pd.read_csv('c:/py_data/py_project/네이버카페-코로나 이후 집콕 정주행.csv', encoding='cp949')
del df_2['Unnamed: 0']
df_2.drop(df_2.index[3], inplace=True)
df_2
```

```python
df_3 = pd.read_csv('c:/py_data/py_project/인스타-코로나 이후 집콕 정주행.csv', encoding='cp949')
del df_3['Unnamed: 0']
df_3
```

```python
df_after = pd.concat([df,df_1,df_2,df_3])
df_after
```

```python
df_after = pd.pivot_table(df_after,
                          index = ['단어'],
                          aggfunc = sum)
df_after
```

```python
df_after.sort_values(by=['횟수'],ascending = False, inplace=True)
df_after
```

```python
df_after.drop(df_after.index[1], inplace=True)
df_after
```

```python
df_after.reset_index(drop=False, inplace=True)
df_after
```

```python
df_after.to_csv('c:/py_data/py_project/After 코로나 단어집계(집콕정주행).csv', encoding='cp949')
```
