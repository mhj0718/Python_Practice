# while문
- 반복해서 문장을 수행해야 할 경우 while문을 사용
- 그래서 while문을 반복문이라고도 부름
- 다음은 while문의 기본 구조

    - while <조건>   
      <수행할 문장1>   
      <수행할 문장2>   
      <수행할 문장3>   
      ...   
- while문은 조건문이 참인 동안에 while문 아래의 문장이 반복해서 수행

```python
i = 1
while i <= 10:
    print(i, '번 출력되었습니다.')
    i += 1
```

```python
# break문
# while문 실행중 break가 실행되면 반복이 중단된다
i = 1
while i <= 10 :
    if i == 7:
        break
    print(i, '번 출력되었습니다.')
    i += 1
```

```python
# continue문 아래의 문장은 실행하지 않고 반복문의
# 사적 자점으로 되돌아간다.
i = 0
while i <= 10:
    i += 1
    if i % 2 == 1:
        continue
    print(i, '번 출력되었습니다.')
```

```python
# continue문 아래의 문장은 실행하지 않고 반복문의
# 시작 지점으로 되돌아 간다
# 무한반복 실행된 i의 값이 정상적으로 증가하지 않기 때문
i = 0
while i <= 10:
    if i % 2 == 1:
        continue
    print(i, '번 출력되었습니다.')
    i += 1
```

```python
stock = 5
while stock > 0:
    money = int(input('돈을 얼마 넣으시겠습니까? : '))
    if money > 500:
        print('상품이 나왔습니다.')
        print('거스름돈은 \\', str(money - 500), '입니다.')
        stock -= 1
        print('재고는 ', stock, '개 입니다.')
    elif money == 500:
        print('상품이 나왔습니다.')
        stock -= 1
        print('재고는 ', stock, '개 입니다.')
    else:
        print('금액이 부족합니다.')
        continue
    if stock == 0:
        print('재고가 없습니다. 프로그램을 종료합니다.')
```

```python
# 1부터 1000사이의 수의 3의 배수 합을 구하기
# 결과 166833
# 3의 배수란 3과 나누었을 때 나머지가 0
```

```python
result = 0
i = 1
while i <= 1000:
    if i % 3 == 0:
        result = result + i
    i += 1
    print(result)
```

# for문(반복문)
- 다음은 for문의 기본 구조   
   
       for 변수 in 순서형 자료:   
           <수행할 문장1>   
           <수행할 문장2>   
           <수행할 문장3>   
           ...

```python
# for 변수 in 순서형 자료:
#    문장
lst = [1,2,3,4,5,3,3,2,3,46,7,8,89.9]
for i in lst:
    print(i,'는 lst에 있어요.')   # lst[i]
```

```python
lst = range(1, 11)

for i in lst:
    if i % 2 == 1:
        continue
    if i == 6:
        break
    print(i,'번 출력 되었습니다.')
```
