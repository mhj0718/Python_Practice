```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import date, timedelta

# 날짜 만들기
YY = int(input('기준 년도를 입력하세요(예:2020):'))
mm = int(input('기준 월을 입력하세요(예:7):'))
dd = int(input('기준 일을 입력하세요(예:20):'))
start_date = date(YY,mm,dd)
period = int(input('크롤링 하고자 하는 기간을 입력해 주세요(예:30): '))
timegap = timedelta(days=period)
start_date = start_date - timegap
start_date = start_date.strftime('%Y-%m-%d')
range_date = pd.date_range(start_date, periods=period, freq='D')


point_lst = []
tlt_lst = []
dt_lst = []
for y in range_date:
    # today.strftime('%Y%m%d') 형식 변환(예 : 2020-01-03 > 20200103)
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date=' + y.strftime('%Y%m%d')
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    title5 = soup.find_all('div', class_ = 'tit5')
    point = soup.find_all('td', class_ = 'point')
    for tit5 in title5:
        temp = tit5.find('a').text
        tlt_lst.append(temp)
    for z in range(len(point)):
        temp_1 = point[z].text
        point_lst.append(temp_1)
    dt_lst = dt_lst + [y] * int(len(point))

# df 만들기
movie_star_df = pd.DataFrame({'date':dt_lst, 'name':tlt_lst,
                               'point':point_lst})
movie_star_df.to_csv('c:/py_data/네이버 영화평점 크롤링.csv')
movie_star_df
```
