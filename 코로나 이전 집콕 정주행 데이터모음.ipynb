```python
import pandas as pd
df = pd.read_csv('c:/py_data/py_project/코로나 이전 집콕.csv', encoding='cp949')
del df['Unnamed: 0']
df
```

```python
df_1 = pd.read_csv('c:/py_data/py_project/구글 뉴스-코로나 이전 집콕.csv', encoding='cp949')
del df_1['Unnamed: 0']
```

```python
df_1
```

```python
df_2 = pd.read_csv('c:/py_data/py_project/네이버카페-코로나 이전 집콕 정주행.csv', encoding='cp949')
del df_2['Unnamed: 0']
df_2
```

```python
df_before = pd.concat([df,df_1,df_2])
df_before
```

```python
df_before = pd.pivot_table(df_before,
                          index = ['단어'],
                          aggfunc = sum)
df_before
```

```python
df_before.sort_values(by=['횟수'],ascending = False, inplace=True)
df_before
```

```python
df_before.drop(df_before.index[4], inplace=True)
df_before
```

```python
df_before.reset_index(drop=False, inplace=True)
df_before
```

```python
df_before.to_csv('c:/py_data/py_project/Before 코로나 단어집계(집콕정주행).csv', encoding='cp949')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import random
img_path = 'c:/py_data/TV.jpg'
cloud_mask = np.array(Image.open(img_path))


wordcloud = WordCloud(font_path="c:/Windows/Fonts/malgun.ttf", 
                      colormap='OrRd',
                      background_color="white", mask=cloud_mask)
wc = wordcloud.generate_from_frequencies(before_cloud)
plt.figure(figsize=(30,40))
plt.imshow(wc)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
