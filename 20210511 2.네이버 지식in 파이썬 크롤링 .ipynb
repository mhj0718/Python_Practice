```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import re

# url 가지고 오기
suburl_lst=[]
main_url = 'https://kin.naver.com/search/list.nhn?query='
page_url = '&page='
keyword = input("네이버 지식인 검색어 입력: ")
max_page = int(input("페이지 입력(Max: 150): "))+1

for page_num in tqdm(range(1,max_page)):
    url = main_url + keyword + page_url + str(page_num)
    html = requests.get(url)
    time.sleep(0.1)
    soup = BeautifulSoup(html.text,'html.parser')
    try:
        ul_tag = soup.find('ul',class_='basic1')
        dt_tag_lst = ul_tag.find_all('dt')
        for i in dt_tag_lst:
            suburl_lst.append(i.find('a')['href'])
    except:
        continue

# 게시글 크롤링
tit_lst = []
con_lst = []
ans_lst = []

for kin_url in tqdm(suburl_lst):
    html_sub = requests.get(kin_url)
    soup_sub = BeautifulSoup(html_sub.text,'html.parser')
    
    try:
        tit_text = soup_sub.find("div",class_='title').text
        tit_text = re.sub('[^ㄱ-ㅣ가-힣 ]','',tit_text)
    except:
        print(kin_url)
        continue
    
    try:
        content_text = soup_sub.find("div",class_='c-heading__content').text
        content_text = re.sub('[^ㄱ-ㅣ가-힣 ]','',content_text)
    except:
        content_text = tit_text
        
    try:
        answer_text_lst = soup_sub.find_all('div',class_='se-module se-module-text').text
        answer_text = ''
        for txt in answer_text_lst:
            answer_text = answer_text + " " + re.sub('[^ㄱ-ㅣ가-힣 ]','',
                                                     txt.text)
            
    except:
        answer_text = ''
        
    tit_lst.append(tit_text)
    con_lst.append(content_text)
    ans_lst.append(answer_text)

df = pd.DataFrame({'title':tit_lst,'content':con_lst,'answer':ans_lst})
```

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re

main_url = 'https://kin.naver.com/search/list.nhn?query='
page_url = '&page='
keyword = '파이썬' #input("네이버 지식인 검색어 입력: ")

suburl_lst=[]
for page_num in range(1,11):
    url = main_url + keyword + page_url + str(page_num)
    html = requests.get(url)
    time.sleep(0.1)
    soup = BeautifulSoup(html.text,'html.parser')
    
    ul_tag = soup.find('ul',class_='basic1')
    dt_tag_lst=ul_tag.find_all('dt')
    for i in dt_tag_lst:
        suburl_lst.append(i.find('a')['href'])

title_list = []
content_list = []
answer_list = []

for hyperlink in suburl_lst:
    html_sub = requests.get(hyperlink)
    soup_sub = BeautifulSoup(html_sub.text,'html.parser')
    
    tit_text = soup_sub.find('div',class_='title').text
    tit_text = re.sub('[^ㄱ-ㅣ가-힣 ]','',tit_text)
    title_list.append(tit_text)
    
    try:
        content_text = soup_sub.find("div",class_='c-heading__content').text
    except:
        content_text = ""
    content_text = re.sub('[^ㄱ-ㅣ가-힣 ]','',content_text)
    content_list.append(content_text)
    
    answer_text_lst = soup_sub.find_all('div',class_='se-module se-module-text')
    temp_txt = ''
    for  txt in answer_text_lst:
        answer_txt = temp_txt + re.sub('[^ㄱ-ㅣ가-힣 ]','',txt.text)
    answer_list.append(answer_txt)

kin_df = pd.DataFrame({'질문제목':title_list,'질문내용':content_list,'답변내용':answer_list})
```
