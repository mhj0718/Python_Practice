```python
# 모범답안

import requests
from bs4 import BeautifulSoup  
import pandas as pd
url = 'https://music.bugs.co.kr/chart'
response_data = requests.get(url)
html = response_data.text                # BeautifulSoup import
soup = BeautifulSoup(html, 'html.parser')

# 순위
rank_lst = soup.find_all("div" ,class_="ranking")
# 제목
title_lst = soup.find_all("p",class_='title')
# 가수
artist_lst = soup.find_all("p" ,class_="artist")

tit_lst = [] 
rak_lst = []
art_lst = []
for i in range(len(rank_lst)):
    rank = rank_lst[i].find('strong').text
    
    title = title_lst[i].find('a').text
    
    artist = artist_lst[i].find('a').text
    
    rak_lst.append(rank)
    tit_lst.append(title)
    art_lst.append(artist)
    
bugs_df = pd.DataFrame({'순위':rak_lst,'제목':tit_lst,'가수':art_lst})
bugs_df = bugs_df.set_index('순위', inplace=True)
bugs_df.to_csv('c:/py_data/벅스탑100크롤링.csv', encoding='utf-8')
```
