```python
driver = webdriver.Chrome('c:/py_data/chromedriver.exe')
url = 'https://play.google.com/store/apps/details?id=net.cj.em.tving&showAllReviews=true'
tit_lst = []

driver.get(url)
time.sleep(1.0)
html =driver.page_source
soup = BeautifulSoup(html, 'html.parser')
review = soup.find_all('span', jsname='bN97Pc')

for i in review:
    title = i.text
    tit_lst.append(title)

review_df = pd.DataFrame({'내용':tit_lst})
review_df
```
