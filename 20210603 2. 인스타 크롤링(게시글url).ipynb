## 코드정리

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import pandas as pd
from tqdm import tqdm
import time
import sys

id_ = '01026778751'
pw= 'agw2539nhh!'

def PAGE_DOWN():
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(3)
    
# 게시물 URL 추출
keyword = input('인스타그램에서 원하는 검색어를 입력하세요: ')
count_down = int(input('스크롤 다운 횟수를 입력하세요(정수입력):'))

driver = webdriver.Chrome('c:/py_data/chromedriver.exe')
driver.get('https://www.instagram.com/')
time.sleep(2)

# 로그인
id_input = driver.find_element_by_css_selector('#loginForm > div > div:nth-child(1) > div > label > input')
id_input.send_keys(id_)
pw_input = driver.find_element_by_css_selector('#loginForm > div > div:nth-child(2) > div > label > input')
pw_input.send_keys(pw)
driver.find_element_by_css_selector('#loginForm > div > div:nth-child(3) > button > div').click()
time.sleep(5)
# a = input('로딩 끝 ?') 딜레이 길어지면 편법으로 이렇게 사용

# 키워드로 검색하기
url = 'https://www.instagram.com/explore/tags/'
driver.get(url+keyword)
time.sleep(5)

body = driver.find_element_by_tag_name('body')
sub_url_lst = []

for down in tqdm(range(count_down),'게시글 url 모음..'):
    PAGE_DOWN()
    contents_lst = body.find_elements_by_class_name('v1Nh3')
    
    for sub_url in contents_lst:
        link = sub_url.find_element_by_tag_name('a').get_attribute('href')
        sub_url_lst.append(link)
    sub_url_lst = list(set(sub_url_lst))
```
