```python
ct = input('지역을 입력해주세요.: ')
Sdate = input('검색 범위 시작날짜를 입력해주세요.(예:yyyy-mm-dd): ')
Edate = input('검색 범위 종료날짜를 입력해주세요.(예:yyyy-mm-dd): ')
new_df = df[(df['주소'].str.contains(ct)) & 
            (df['시작일자'] >= Sdate) & 
            (df['종료일자'] <= Edate)]
lat = new_df.iloc[0,8]
lon = new_df.iloc[0,9]
center = [lat,lon]

m = folium.Map(location=center, zoom_start=11)
for i in new_df.index:
    lat1 = new_df.loc[i,'위도']
    lon1 = new_df.loc[i,'경도']
    folium.Marker(location=[lat1, lon1],
                  icon=folium.Icon(color='blue'),
                  popup=new_df.loc[i,'축제명']).add_to(m)
m
```

```python
m.save('c:/py_data/maptest2.html')
```
